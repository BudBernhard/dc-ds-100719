```python
import nltk
import requests
response = requests.get('https://raw.githubusercontent.com/learn-co-students/dc-ds-100719/master/Warm-Up-Files/alice_in_wonderland.txt')
data = response.text
```

```python
dict_alice_words = {} 
pattern = "([a-zA-Z]+(?:'[a-z]+)?)"
alice = nltk.regexp_tokenize(data, pattern)
for n in alice:
    if n in dict_alice_words.keys():
        dict_alice_words[n]+=1 
    else: 
        dict_alice_words[n] = 1
print(dict_alice_words)
```

```python
from collections import Counter
cnt = Counter(dict_alice_words)
print(cnt.most_common(10))
```
